### Import the file from excel and save to a dataframe

```python
import pandas as pd

df = pd.read_excel (r'C:\Users\Tijesunimi2-PC\Documents\10academy_Telecommunication\Week2_challenge_data_source.xlsx')
df
```

### Fill all missing values in the dataset with the mean

```python
df.fillna(df.mean())
```

#### Task 1.1 - Aggregate the Number of xDR sessions,  Session duration,  the total download (DL) and upload (UL) data , the total data volume (in Bytes) during this session for each application (Social Media, Google, Email, YouTube, Netflix, Gaming).

#### Aggregate the Number of xDR sessions

```python
df.groupby('MSISDN/Number')['Bearer Id'].count()
```

#### Aggregate the Session duration

```python
df.groupby('MSISDN/Number')['Dur. (ms)'].sum()
```

#### Aggregate the total download (DL) and upload (UL) data

#### To do this, we first need to fill the last missing row of DL and UL

```python
df.groupby('MSISDN/Number')['Total DL (Bytes)','Total UL (Bytes)'].sum()
```

#### Average of the two columns "total download (DL) and upload (UL) data"

```python
df['TotalDLandULaverage'] = df[['Total DL (Bytes)','Total UL (Bytes)']].mean(axis=1)
df['TotalDLandULaverage']
```

#### Aggregate the total data volume (in Bytes) during this session for each application (Social Media, Google, Email, YouTube, Netflix, Gaming).

```python
df['TotalSocialMediaDataVolume'] = df[['Social Media DL (Bytes)','Social Media UL (Bytes)']].mean(axis=1)
df['TotalGoogleDataVolume'] = df[['Google DL (Bytes)','Google UL (Bytes)']].mean(axis=1)
df['TotalEmailDataVolume'] = df[['Email DL (Bytes)','Email UL (Bytes)']].mean(axis=1)
df['TotalYoutubeDataVolume'] = df[['Youtube DL (Bytes)','Youtube UL (Bytes)']].mean(axis=1)
df['TotalNetflixDataVolume'] = df[['Netflix DL (Bytes)','Netflix UL (Bytes)']].mean(axis=1)
df['TotalGamingDataVolume'] = df[['Gaming DL (Bytes)','Gaming UL (Bytes)']].mean(axis=1)

concatenate = pd.concat([df['TotalSocialMediaDataVolume'], df['TotalGoogleDataVolume'], df['TotalEmailDataVolume'],
                        df['TotalYoutubeDataVolume'], df['TotalNetflixDataVolume'], df['TotalGamingDataVolume']], axis=1)
concatenate
```

#### Start by identifying the top 10 handsets used by the customers.

#### Then, identify the top 3 handset manufacturers

#### Getting the minimum value, the maximum value, the average, the 1st, 2nd & 3rd quartile for the Number of xDR sessions

```python
print("Minimum", df.groupby('MSISDN/Number')['Bearer Id'].count().sort_values(ascending=True).head(1))
print("")
print("Maximum", df.groupby('MSISDN/Number')['Bearer Id'].count().sort_values(ascending=False).head(1))
print("")
print("Average", df.groupby('MSISDN/Number')['Bearer Id'].count().mean())
print("")
df.groupby('MSISDN/Number')['Bearer Id'].count().quantile([0.25, 0.50, 0.75], interpolation='nearest')
```
